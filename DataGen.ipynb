```python
from reliability.Distributions import Weibull_Distribution
import matplotlib.pyplot as plt
from cycler import cycler
plt.style.use('ggplot')

import pandas as pd
```

```python
char_strengts = [856.2, 860.4, 873.5, 836.2, 843.3]
moduli =        [9.1,   10.1,  12.1,  9.5, 10.5]
batch = ['A', 'B', 'C', 'D', 'E']

fig, ax = plt.subplots(dpi=300, figsize=(4, 3))
for i in range(len(char_strengts)):
    dist = Weibull_Distribution(alpha=char_strengts[i], beta=moduli[i])
    dist.PDF(lw=3.1, color='k')
    dist.PDF(label=f'Batch {batch[i]}', lw=2)
    samples = dist.random_samples(30)
    # samples = pd.Series(samples).to_csv(f'data/batch_{i}_strengths.csv', index=False, header=False)
plt.legend( loc='upper left', fontsize=10, facecolor='white', edgecolor='black', framealpha=1)
plt.title('')
plt.xlabel('Strength [MPa]')
plt.ylabel('Probability Density')
plt.xlim(400, 1100)
plt.show()
```
